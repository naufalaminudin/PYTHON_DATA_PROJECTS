```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_US = df[df['job_country'] == 'United States'].copy()

# Menambahkan kolom 'job_posted_month' berdasarkan bulan dari 'job_posted_date'
df_US['job_posted_month'] = df['job_posted_date'].dt.strftime('%B')

# Membuat pivot table berdasarkan bulan dan job_title_short
df_US_pivot = df_US.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')

# Mengatur ulang index untuk menyimpan bulan sebagai kolom
df_US_pivot.reset_index(inplace=True)

# Menambahkan kolom 'job_posted_month_no' sebagai nomor bulan
df_US_pivot['job_posted_month_no'] = pd.to_datetime(df_US_pivot['job_posted_month'], format='%B').dt.month

# Mengurutkan DataFrame berdasarkan nomor bulan
df_US_pivot.sort_values('job_posted_month_no', inplace=True)

# Mengatur kembali index berdasarkan 'job_posted_month'
df_US_pivot.set_index('job_posted_month', inplace=True)

# Menghapus kolom 'job_posted_month_no' yang tidak lagi diperlukan
df_US_pivot.drop(columns='job_posted_month_no', inplace=True)

# Menampilkan hasil
df_US_pivot
```

```python
df_software_pivot = pd.read_csv('https://lukeb.co/software_csv', index_col='job_posted_month')

df_software_pivot
```

```python
df_US_merged = df_US_pivot.merge(df_software_pivot, on='job_posted_month')
```

```python
top_5 = df_US_merged.sum().sort_values(ascending=False).head().index.to_list()

top_5
```

```python
df_US_merged[top_5].plot(kind="line")
plt.title('Monthly Job Postings for Top Tech Jobs in the US')
plt.xlabel('2023')
plt.ylabel('Job Counts')
plt.ylim(0, 20000)
plt.legend()
plt.show()
```
