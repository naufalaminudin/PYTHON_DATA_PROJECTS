```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## Counts of Job Title

```python
df.pivot_table(index = 'job_title_short', aggfunc='size')
```

```python
df.pivot_table(values ='salary_year_avg', index = 'job_title_short', aggfunc='median')
```

```python
df.pivot_table(values ='salary_year_avg', index = 'job_country', columns= 'job_title_short', aggfunc='median')
```

```python
top_countries = df['job_country'].value_counts().head(6).index

top_countries
```

```python
df_job_country_salaries = df.pivot_table(
    values='salary_year_avg',
    index='job_country',
    columns='job_title_short',
    aggfunc='median'
)
```

```python
df_job_country_salaries = df_job_country_salaries.loc[top_countries]
```

```python
job_titles = ['Data Analyst', 'Data Scientist', 'Data Engineer']

df_job_country_salaries = df_job_country_salaries[job_titles]

df_job_country_salaries
```

```python
df_job_country_salaries.plot(kind='bar')
plt.ylabel('Median Salary')
plt.xlabel('')
plt.title('Median Salary by Country and Job Titles')
plt.xticks(rotation=45, ha='right')
plt.show()
```
